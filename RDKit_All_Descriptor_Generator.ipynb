```python
## import the necessary modules
import pandas as pd
import rdkit
from rdkit import Chem
from rdkit.Chem import Descriptors
from rdkit.Chem.Descriptors import *
```

```python
descriptors = Descriptors.descList
```

```python
descriptors = pd.DataFrame(descriptors)
descriptors.columns = ["Descriptor_Name","RDKit_Fun" ]
descriptors
```

```python
descriptors_names = descriptors["Descriptor_Name"].values
```

```python
descriptors_names
```

```python
## load the sample dataframe
sample = pd.read_csv("sample.csv")
```

```python
sample.head(10)
```

```python
sample_mod = sample[["ChEMBL ID", "Smiles"]]
```

```python
sample_mod.head(5) ## modified sample dataframe looks ike this---
```

```python
### Let's generate all the descriptors
for prop in descriptors_names:
  df = [eval(prop)(Chem.MolFromSmiles(smi)) for smi in sample_mod["Smiles"].values]
  df = pd.DataFrame(df)
  df.columns=[prop]
  sample_mod = pd.concat([sample_mod,df], axis=1)
```

```python
sample_mod.head(10)
```

```python
sample_mod.to_csv("sample_descriptors.csv", index=None)
```
